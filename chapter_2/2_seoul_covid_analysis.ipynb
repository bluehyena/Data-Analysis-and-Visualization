```python
# 월만 추출하기

df["월"] = df["확진일자"].dt.month
df[["확진일","확진일자","월"]].tail()
```

## 선 그래프에 값 표시하기

```python
df["확진일자"]
```

```python
df["월일"] = df["확진일자"].astype(str).map(lambda x : x[-5:])
day_count = df["월일"].value_counts().sort_index()
df["월일"].value_counts().sort_index().plot()
day_count.iloc[2]
```

```python
g = day_count.plot(figsize=(15,4))
for i in range(len(day_count)):
    case_count = day_count.iloc[i]
    g.text(x=i, y=case_count, s=case_count)
```

```python
g = day_count.plot(figsize=(15,4))
for i in range(len(day_count)):
    case_count = day_count.iloc[i]
    if case_count > 25:        
        g.text(x=i, y=case_count, s=case_count)
```

## 최근 데이터만 슬라이싱으로 나눠 그리기

```python
h = day_count[-50:].plot.bar(figsize=(15,4))
h.axhline(day_count.median(), linestyle=":", color="red")
# h.axhline(8, linestyle=":", color="red")

for i in range(50):
    case_count = day_count[-50:].iloc[i]
    if case_count > 20:        
        h.text(x=i, y=case_count, s=case_count)
```
