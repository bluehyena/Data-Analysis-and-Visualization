## area plot

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(np.random.rand(10, 4), columns = ['a','b','c','d'])
df.head()
```

```python
df.plot.area() # 비어있는 값은 0으로 채워짐, df.dropna() 나 df.fillna()로 전처리 권장
```

```python
df.plot() # stacked 가 default 값임
```

```python
df.plot.area(stacked=False, grid=True) # grid == 격자
```
