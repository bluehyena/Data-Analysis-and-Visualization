```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
columns = ['sepal_width','sepal_length','species']
df[columns].head()
```

```python
df['sepal_width'] # 추천 방법
```

```python
df.sepal_width # 한글이나 특수문자는 불가
```

```python
df.filter(regex='regex') # 정규표현식
```

```python
df[columns].filter(regex='_')
```

```python
df.filter(regex='length$') # $ 로 끝나는 것을 불러옴
```

```python
df[columns].filter(regex='^sepal') # ^ 로 시작하는 것을 불러옴
```

```python
df[columns].filter(regex='^se') # ^ 로 시작하는 것을 불러옴
```

```python
df.filter(regex='^(?!species$).*')
```
