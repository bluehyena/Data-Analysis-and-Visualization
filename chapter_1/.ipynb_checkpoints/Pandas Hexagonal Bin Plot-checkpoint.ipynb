# Hexagonal Bin Plot
* Hexbin plot 은 데이터가 클 때 각각의 점을 산점도(scatter plot)로 표현할 때의 단점을 보완할 수 있다.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl

%matplotlib inline

mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
```

```python
df.tail()
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25)
```

```python
df.plot.hexbin(x='a', y='b', gridsize=10)
```

```python
df['z'] = np.random.uniform(0,3,1000)
df['z']
```

```python
df.plot.hexbin(x='a', y='b', C='z', reduce_C_function = np.max, gridsize=25)
```

* reduce_C_function
    - 기본적으로 각 (x,y) 점 주변의 개수에 대한 히스토그램이 계산된다. C 및 reduce_C_function 인수에 값을 전달하여 대체 집계를 지정할 수 있다. C는 각(x,y) 점에서 값을 지정하고 reduce_C_function은 빈의 모든 값을 단일 숫자 (ex : 평균, 최대 값, 합계, 표준)으로 줄이는 하나의 인수의 함수이다. 이 예에서 위치는 열(column) a or b에 의해 주어지며 값은 열(z)에 의해 주어진다. 빈은 numpy의 max 함수로 집계된다.

```python
df.plot.hexbin(x='a', y='b', C='z', reduce_C_function = np.min, gridsize=35)
```
